```python
#évolution du nombre de personnes absolues
import pandas as pd
import numpy as np

evol_maj = ouvrir_json("evol_nb_autricesabs.json")

x = list(evol_maj.keys())
h = []
for subl in list(evol_maj.values()):
    if len(subl) > 0:
        h.append(subl[0])
    else:
        h.append(0)
        
f = []
for subl in list(evol_maj.values()):
    if len(subl) > 1:
        f.append(subl[1])
    else:
        f.append(0)

y = []
for i in range(len(h)):
    el_f = f[i]
    el_h = h[i]
    if el_f+el_h> 0:
        y.append(el_f/(el_h+el_f))
    else:
        y.append(np.nan)
        #y.append(0)


#pour que les valeurs pas remplies ne soient pas = 0       
filled = pd.Series(y).fillna(limit=2, method='ffill')

# Plot a simple line chart
plt.plot(x, filled, 'r', label="Proportion d'autrices")
#plt.plot(x, y, 'r', label="Proportion d'autrices")

plt.legend(loc="upper right")
plt.xticks(rotation=90, ha='right')
a = plt.gca()
  
# set visibility of x-axis as False
xax = a.axes.get_xaxis()
for label in xax.get_ticklabels()[::2]:
    label.set_visible(False)
    
plt.savefig("evolution_nbautrices_proportion_fillna.png",bbox_inches='tight')
plt.show()
```
