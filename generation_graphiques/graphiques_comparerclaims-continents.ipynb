```python
import json
def ouvrir_json(chemin):
  f = open(chemin, encoding="utf-8")
  toto = json.load(f)
  f.close()
  return toto

import numpy as np
import matplotlib.pyplot as plt
```

```python
# REGARDER L'EVOLUTION DU NB DE CONTINENTS DANS LE TEMPS
evol_maj = ouvrir_json("evol_nb_conti_corr.json")
# Define data values
x = list(evol_maj.keys())
asie = []
af = []
eu = []
na = []
sa = []
oc = []
for year, dico in evol_maj.items():
    #for conti, nb in dico.items():
    asie.append(dico["as"])
    af.append(dico["af"])
    eu.append(dico["eu"])
    na.append(dico["na"])
    sa.append(dico["sa"])
    oc.append(dico["oc"])

# Plot a simple line chart
plt.plot(x, asie, label="Asie")
# Plot another line on the same chart/graph
plt.plot(x, af, label="Afrique")
plt.plot(x, eu, label="Europe")
plt.plot(x, na, label="Am Nord")
plt.plot(x, sa, label="Am Sud")
plt.plot(x,oc,label="Océanie")

plt.legend(loc="upper left")
plt.xticks(rotation=90, ha='right')
a = plt.gca()
  
# set visibility of x-axis as False
xax = a.axes.get_xaxis()
for label in xax.get_ticklabels()[::2]:
    label.set_visible(False)
    
plt.savefig("stopwords/corr/evolution_conti.png",bbox_inches='tight')
plt.show()
plt.clf()
```
