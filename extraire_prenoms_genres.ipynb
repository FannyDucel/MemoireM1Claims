```python
import json
import warnings
import pandas as pd
import glob
import re
warnings.filterwarnings('ignore')

def ecrire_json(chemin, contenu):
  w = open(chemin, "w", encoding="utf-8")
  w.write(json.dumps(contenu, indent=2, ensure_ascii=False))
  w.close()

def ouvrir_json(chemin):
  f = open(chemin, encoding="utf-8")
  toto = json.load(f)
  f.close()
  return toto 

def lire_fichier(chemin):
    with open(chemin,encoding="utf-8",errors="ignore") as f:
        text = f.read()
    return text
```

```python
"""Commencer par extraire les prénoms des auteur-ices"""
dico_auteurices = {}
# Transformer les csv de bib en un gros dictionnaire de dict type {annee : {url:auteur, url:auteur}}
for chemin in glob.glob("acl_bib/*.csv"):
    chemin_p = chemin.split(".")
    annee = re.sub("acl_bib\\\\","",chemin_p[0])
    df = pd.read_csv(chemin, index_col=0)
    df = df.fillna("x")
    dico_auteurs = df.to_dict()
    for author, dico in dico_auteurs.items():
        dico_auteurs = dico
        dico_auteurices[annee]=dico
        
for annee, dico in dico_auteurices.items():
    for url, liste_auteurs in dico.items():
        liste_auteurs = re.sub("and\\n", " ", liste_auteurs)
        dico_auteurices[annee][url]=liste_auteurs
```

```python
"""Approche si on prend le genre majoritaire du groupe"""
dico_genre_auteur = ouvrir_json("dico_genre_auteur_combinedv2.json")
dico_nb_genre_url = {}
for url, dico_genre in dico_genre_auteur.items():
    nb_h = 0
    nb_f = 0
    for prenom, genre in dico_genre.items():
        if genre == "male":
            nb_h+=1
        else: #if female
            nb_f+=1
    if nb_h > nb_f :
        genre = "h"
    elif nb_f > nb_h:
        genre = "f"
    else: #en cas d'égalité
        genre = "e"
    if url not in dico_nb_genre_url:
        dico_nb_genre_url[url] = genre
        
print(dico_nb_genre_url)
ecrire_json("dico_nb_genre_urlv06-2.json",dico_nb_genre_url)

#constituer une liste avec les url d'articles qui ont été écrits maj par hommes, 1 pour femmes, 1 pour égalité
liste_url_h, liste_url_f, liste_url_e = [],[],[]
for url, genre in dico_nb_genre_url.items():
    #transformer les url en chemin en remplaçant http...org > ../../Stage_L3/aan/papers_text\\ au début et .txt à la fin
    url = url.split("/")
    #url = "../../Stage_L3/aan/papers_text\\"+url[-1]+".txt"
    url = url[-1]+".txt"
    if genre == "h":
        liste_url_h.append(url)
    if genre =="f":
        liste_url_f.append(url)
    if genre =="e":
        liste_url_e.append(url)
print(len(liste_url_h),len(liste_url_f),len(liste_url_e))
ecrire_json("liste_url_hv06-2.json",liste_url_h)
ecrire_json("liste_url_fv06-2.json",liste_url_f)
ecrire_json("liste_url_ev06-2.json",liste_url_e)

#prendre en compte les années pour voir l'évolution du nombre d'autrices dans le temps
dico_nb_genre_url = ouvrir_json("dico_nb_genre_urlv06-2.json")
nb_art_h = 0
nb_art_f = 0
nb_art_e = 0
dico_annees = {}
[dico_annees.setdefault(i, []) for i in range(1979,2021)] 
for url, genre in dico_nb_genre_url.items():
    if "2020.acl" in url:
        if genre=="h":
            nb_art_h+=1
        elif genre=="f":
            nb_art_f+=1
            #print(url)
        else:
            nb_art_e+=1
        dico_annees[2020]=[nb_art_h,nb_art_f,nb_art_e]
        
    for an in range(1979,2020):
        if "P"+str(an)[-2:] in url:
            if genre=="h":
                nb_art_h+=1
            elif genre=="f":
                nb_art_f+=1
                #print(url)
            else:
                nb_art_e+=1
            dico_annees[an]=[nb_art_h,nb_art_f,nb_art_e]
        #print(url)
print(dico_annees)
ecrire_json("evol_nb_autrices.json",dico_annees)
```
